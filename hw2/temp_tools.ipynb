```python
def geo_mean(iterable):
    a = np.array(iterable)
    return round(a.prod()**(1.0/len(a)), 3)
```

```python
BENGMARKS_BASE_LINE = [12.84855, 7.86705, 2.9598, 2.48055, 3.2055, 34.48905, 4.4718, 4.14915]
result = {"name": [], 'config': [], 'D_miss_rate': [], 'MP': [], 'ideal_CT': [], 'CT': [], 'eval_factor': []} # config eg: 'capacity---entries---set_ways---line_size_in_B'
eval_result = {'config': [], 'geo_mean': []}
li_of_eval_factor = []

benchmarks = inp.strip().split('end\n')[:16]
# print(benchmarks)
# print(benchmarks)
# benchmarks.pop() # remove last empty string
for i, benchmark in enumerate(benchmarks):
    if benchmark == '':
        continue
    
    lines = benchmark.strip().split('\n')
    
    temp = lines[0].split('-')
    name, capacity, entries, set_ways, line_size_in_B = temp
    capacity = f'{int(capacity)//1024}KB' if int(capacity) <= 524288 else f'{int(capacity)//1048576}MB' # transfer capacity from B to KB or MB
    config = '-'.join([capacity, entries, set_ways, line_size_in_B])
    # if i % 6 == 0:
    D_miss_rate = float(re.split(' {2,}', lines[9])[1][:-2]) * 0.01
    # print(lines[9])
    if D_miss_rate > 1:
        D_miss_rate = float(re.split(' {2,}', lines[10])[1][:-2]) * 0.01
    # print(D_miss_rate)
    MP = 100 + (int(temp[4])//4-1)*5
    ideal_CT = float(df.loc[set_ways, capacity] if line_size_in_B == '16' else df2.loc[set_ways, capacity])    # read given table to get CT
    CT = ideal_CT if ideal_CT > 0.6 else 0.6
    
    result['name'].append(name)
    result['config'].append(config)
    result['D_miss_rate'].append(D_miss_rate)
    result['MP'].append(MP)
    result['ideal_CT'].append(ideal_CT)
    result['CT'].append(CT)
    
    eval_factor = (D_miss_rate * MP + CT) / BENGMARKS_BASE_LINE[i % 8]
    result['eval_factor'].append(round(eval_factor, 3))
    
    # init new 8 benchmarks
    if i % 8 == 0:
        eval_result['config'].append(config)
        li_of_eval_factor = []
    
    # record the eval_factor for geo mean calculation
    li_of_eval_factor.append(eval_factor)
    # if i == 7:
    #     print("+=======")
    #     print(D_miss_rate)
    #     print(MP)
    #     print(CT)
    #     print(BENGMARKS_BASE_LINE[i % 8])
    
    # use every 8 benchmarks to calculate geo_mean
    if i % 7 == 0 and i:
        eval_result['geo_mean'].append(geo_mean(li_of_eval_factor))

result = pd.DataFrame(result)
eval_result = pd.DataFrame(eval_result)
print(result)
print(eval_result)
```
